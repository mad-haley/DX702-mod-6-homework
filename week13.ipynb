```python
import pandas as pd
import statsmodels.api as sm
from linearmodels.iv import IV2SLS

# load the dataset
df = pd.read_csv('homework_13.1.csv')

df.head()
```

```python
# define variables
Z = df['Z']
X = df['X']
Y = df['Y']   # instrument

# first stage
X1 = sm.add_constant(Y)
fs = sm.OLS(X, X1).fit()
df['Xhat'] = fs.fittedvalues

# second stage
X2 = sm.add_constant(df['Xhat'])
ss = sm.OLS(Z, X2).fit()
ss.params
```

question 2

```python
df = pd.read_csv("homework_13.2.csv")

# define variables
Y2 = df["Y2"]
X2 = df["X2"]
Z2 = df["Z2"]  # instrument

# first stage
X1 = sm.add_constant(Z2)
fs = sm.OLS(X2, X1).fit()
df["Xhat"] = fs.fittedvalues

# second stage
X2_stage2 = sm.add_constant(df["Xhat"])
ss = sm.OLS(Y2, X2_stage2).fit()

ss.params
```

```python
df = pd.read_csv("homework_13.3.csv")
df.head()
```

```python
# create interaction term
df["ZW_int"] = df["Z3"] * df["W3"]

# linear regression of X3 on Z3, W3, ZW_int
X = sm.add_constant(df[["Z3", "W3", "ZW_int"]])
y = df["X3"]

model = sm.OLS(y, X).fit()
print(model.summary())

print(model.params)
```
