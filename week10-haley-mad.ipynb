question 1

```python
import statsmodels.formula.api as smf

df2 = df.copy()
# treat time as categorical
model = smf.ols('y ~ X + C(time)', data=df2).fit()
fe = model.params.filter(like='C(time)')
# Need baseline intercept + dummies; construct full FE per time
intercept = model.params['Intercept']
effects = {}
for t in sorted(df2.time.unique()):
    key = f'C(time)[T.{t}]'
    effects[t] = intercept + (model.params[key] if key in model.params else 0)
effects
```

question 2

```python
model_city = smf.ols('y ~ X + C(city)', data=df).fit()
intercept = model_city.params['Intercept']
effects_city = {}
for c in sorted(df.city.unique()):
    key = f'C(city)[T.{c}]'
    effects_city[c] = intercept + (model_city.params[key] if key in model_city.params else 0)
effects_city
```

question 3

```python
import numpy as np
import statsmodels.api as sm


num = 10000
X = np.clip(np.random.normal(3, 1, (num,)), 0.01, 100)
Z = np.clip(np.random.normal(3, 1, (num,)), 0.01, 100)
Y = np.log(X + Z) + np.random.normal(0, 1, (num,))

# model exp(Y) as function of X and Z
expY = np.exp(Y)

df = pd.DataFrame({'expY': expY, 'X': X, 'Z': Z})
Xmat = sm.add_constant(df[['X', 'Z']])
model = sm.OLS(df['expY'], Xmat).fit()
model.summary()
```
